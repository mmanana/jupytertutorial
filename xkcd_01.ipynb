# Ejemplos de uso de Jupyter

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
x = np.linspace(0, 2000, N)
y = np.minimum((x > 500) * ((x - 500)/50), 10.0*np.ones(N))
```

```python
with plt.xkcd():

    fig = plt.figure()

    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_ylim([0, 10.5])
    
    ax.annotate(
        '500 h: POR FIN ENTIENDO ALGO!!',
        xy=(500, 0), arrowprops=dict(arrowstyle='->'), xytext=(700, 1))

    ax.annotate(
        '750 h: POR FIN ENTIENDO LO SUFICIENTE!!',
        xy=(750, 5), arrowprops=dict(arrowstyle='->'), xytext=(1000, 3))

    ax.annotate(
        '1000 h: SOY UN GENIO!!',
        xy=(1000, 10), arrowprops=dict(arrowstyle='->'), xytext=(1300, 6))
    
    
    ax.plot(x,y)
    ax.set_xlabel('Horas de estudio')
    ax.set_ylabel('Nota final')
    fig.text(
            0.5, -0.1,
            '"Regla básica de estudio" utilizando xkcd por Mario Mañana',
            ha='center')
```
